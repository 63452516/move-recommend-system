```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv(r'C:\Users\acer\Downloads\movies\TMDB_movies.csv')
print(df)
```

```python
df= df[['id','title','overview','genre']]
```

```python
df
```

```python
df['tags']=df['overview']+df['genre']
```

```python
df
```

```python
movies = df.drop(columns=['overview','genre'])
```

```python
movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vec=cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
s=cosine_similarity(vec)
```

```python
s
```

```python
movies[movies['title']=="The Godfather"].index[0]
```

```python
dist= sorted(list(enumerate(s[2])), reverse=True , key = lambda vec:vec[1])
for i in dist[0:5]:
    print (movies.iloc[i[0]].title)
```

```python
def recommand(df):
    index=movies[movies['title']==df].index[0]
    dist= sorted(list(enumerate(s[index])), reverse=True , key = lambda vec:vec[1])
    for i in dist[0:5]:
        print (movies.iloc[i[0]].title)
```

```python
recommand("Zombieland")
```
